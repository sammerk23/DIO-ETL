# Teste ETL

```python
import pandas as pd

df = pd.read_csv('/content/products-1000.csv')
df.head()
```

```python
print(df.isna().sum())
```

```python
filteredData = df[(df['Price'] > 100) & (df['Currency'] == 'USD')]
filteredData.head()
```

```python
filteredData.count()
```

```python
print(filteredData['Size'].unique())
```

```python
filteredData.loc[:, 'Size'] = filteredData['Size'].replace('M', 'Medium')
filteredData.loc[:, 'Size'] = filteredData['Size'].replace('S', 'Small')
filteredData.loc[:, 'Size'] = filteredData['Size'].replace('L', 'Large')
filteredData.loc[:, 'Size'] = filteredData['Size'].replace('XL', 'Extra Large')
```

```python
print(filteredData['Size'].unique())
```

```python
filteredData.to_csv('filteredData.csv', index=False)
```
